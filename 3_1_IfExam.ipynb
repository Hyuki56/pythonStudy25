```python
# if 문은 제어문으로 참, 거짓을 판단하여 흐름을 조절함

# ex) 돈이 있으면 택시를 타고가고, 없으면 걸어가라.
money = True
if money:
    print("택시를 타고갑니다")
else:
    print("걸어갑니다.")
print("학교에 도착하였습니다.")
```

```python
money = int(input("현재 얼마있나요??")) # input으로 입력받은 값은 str타입임.
# print("현재 주머니에 %d 가 있네요" %money)

if money >= 3000:
    print("택시를 타고 감")
else:
    print("걸어감")
print("학교에 도착")
```

```python
money = 2000
card = True
if money >= 3000 | card:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```
